```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors
import matplotlib.pyplot as plt

# Test 5 molecules
smiles = ['CCO', 'c1ccccc1', 'CC(=O)O', 'CN1C=NC2=C1C(=O)N(C(=O)N2C)C', 'CC(C)Cc1ccc(cc1)C(C)C(=O)O']
names = ['Ethanol', 'Benzene', 'Acetic Acid', 'Caffeine', 'Ibuprofen']

for smi, name in zip(smiles, names):
    mol = Chem.MolFromSmiles(smi)
    print(f"\n{name} ({smi})")
    print(f"  MW: {Descriptors.MolWt(mol):.2f}")
    print(f"  LogP: {Descriptors.MolLogP(mol):.2f}")
    
    img = Draw.MolToImage(mol)
    plt.figure(figsize=(4,3))
    plt.imshow(img)
    plt.title(name)
    plt.axis('off')
    plt.show()

print("\n RDKit working!")
```

```python
# 3. Calculate properties
mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
tpsa = Descriptors.TPSA(mol)

print(f"Molecular Weight: {mw:.2f}")
print(f"LogP: {logp:.2f}")
print(f"TPSA: {tpsa:.2f}")
```

```python


# 4. Try 10 different molecules
smiles_list = [
    'CC',           # Ethane
    'CCO',          # Ethanol
    'c1ccccc1',     # Benzene
    'CC(=O)O',      # Acetic acid
    'CN1C=NC2=C1C(=O)N(C(=O)N2C)C',  # Caffeine
    'CC(C)Cc1ccc(cc1)C(C)C(=O)O',    # Ibuprofen
    'C1=CC=C(C=C1)O',                 # Phenol
    'CCN(CC)CC',                      # Triethylamine
    'C1CCCCC1',                       # Cyclohexane
    'CCCCCCCCCC'                      # Decane
]

for i, smi in enumerate(smiles_list):
    mol = Chem.MolFromSmiles(smi)
    if mol:
        print(f"\n{i+1}. {smi}")
        print(f"   MW: {Descriptors.MolWt(mol):.2f}")
        print(f"   LogP: {Descriptors.MolLogP(mol):.2f}")
        # Visualize
        img = Draw.MolToImage(mol)
        plt.figure(figsize=(3,3))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Molecule {i+1}")
        plt.show()
```
